```python
import pandas as pd

df = pd.read_csv("../data/raw/UNSW_NB15_training-set.csv")
df = df.dropna()
df.head()
```

```python
df['is_attack'] = df['label'].apply(lambda x: 1 if x == 1 else 0)
```

```python
window_size = 500

attack_counts = df['is_attack'].groupby(df.index // window_size).sum()
```

```python
ts = pd.DataFrame({
    'time': range(len(attack_counts)),
    'attacks': attack_counts.values
})

ts.head()
```

```python
ts.to_csv("../data/processed/unsw_attack_timeseries.csv", index=False)
```
